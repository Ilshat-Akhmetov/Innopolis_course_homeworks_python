```python
import matplotlib.pyplot as plt 
import numpy as np 
import seaborn as sns
import pandas as pd
import plotly.express as px
import matplotlib.image as mpimg
from PIL import Image
```

```python
step = 0.01
start = 0
stop = 4
```

```python
x_array = np.arange(start=start, stop=stop+step, step=step)
f = lambda x: np.cos(20*x)/(x+0.1)
y_array = f(x_array)
```

```python
plt.plot(x_array, y_array)
```

```python
sns.lineplot(x=x_array, y=y_array)
```

```python
px.line(x=x_array, y=y_array)
```
