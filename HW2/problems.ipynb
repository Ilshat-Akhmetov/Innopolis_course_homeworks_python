```python
import numpy as np
import pandas as pd
```

```python
def create_matrix(rows: int = 8, cols: int = 8):
    matrix = np.zeros((rows, cols), dtype=np.int8)
    matrix[::2,1::2]=1
    matrix[1::2,::2]=1
    return matrix
```

```python
assert (create_matrix(2, 2) == np.array([[0, 1],[1, 0]])).all()
```

```python
create_matrix()
```

```python
n = 2000
mean_v = 0
var = 5
arr = np.random.normal(mean_v, np.sqrt(var), n)
emp_mean = np.mean(arr)
emp_var = np.var(arr)
print(f"экспериментальное среднее: {emp_mean} дисперсия {emp_var}")
```
